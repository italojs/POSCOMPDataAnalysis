```python
import textdistance

cosine = textdistance.cosine

class Comparison():
    def __init__(self, q1, q2):
        self.distance = float('%.2f' % cosine.distance(q1.text, q2.text))
        self.similarity = float('%.2f' % cosine.similarity(q1.text, q2.text))
        self.comparedQuestion = q2
        
class Question(object):
    def __init__(self, PDFID, index, text):
        self.ID = f'{PDFID}_q{index}'
        self.PDFID = PDFID
        self.text = text
        self.comparisons = []
        
    def compare(self, question):
        if self.PDFID is not question.PDFID:
            self.comparisons.append(Comparison(self, question))
```

```python
import re
from tika import parser as pdfReader
    
def _cleanPDF(content):
    text = content \
        .replace('\n', '') \
        .replace('\r', '') \
        .replace('‚ñ¨','') \
        .replace('Execu√ß√£o: Fundatec', '')
    
    regexs = [r'(EXAME POSCOMP)\s\d{4}']
    
    for rgx_match in regexs:
        new_text = re.sub(rgx_match, '', text)
    return text

def _createQuestions(PDFID, content):
    questions = []
    splitedQuestions = content.split('QUEST√ÉO')
    # Removing alternatives
    for i, text in enumerate(splitedQuestions[1:]):
        questionText = text.split('A)')[0]
        question = Question(PDFID, i, questionText)
        questions.append(question)
    return questions
    

class PDF(object):
    def __init__(self, path):
        raw = pdfReader.from_file(path)
        self.content = _cleanPDF(raw['content'])
        
        self.ID = path
        self.name = path.split('/')[-1]
        self.questions = _createQuestions(self.ID, self.content)
```

```python
import glob

questions = []
for path in glob.glob('./Provas-POSCOMP/**/caderno_*.pdf'):
    pdf = PDF(path)
    questions = [*questions, *pdf.questions]
```

```python
for question in questions:
    for c in question.comparisons:
        if c.similarity >= .80 and c.similarity < .86:
            print(f'PDFID: {question.PDFID}\n')
            print(f'text: {question.text}\n\n')

            print(f'PDFID: {c.comparedQuestion.PDFID}\n')
            print(f'comparedText: {c.comparedQuestion.text}\n\n')

            print(f'distance: {c.distance}\n')
            print(f'similarity: {c.similarity}\n') 
            print('-'*60)
```
